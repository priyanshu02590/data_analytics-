```python
3.Comparison operators(==,!=,<=,>=,<,>)
```

```python
x=17
y=13
print(x<y)
print(x<=y)
print(x>y)
print(x>=y)
print(x!=y)
print(x==y)
```

```python
4.Logical operators(And,Or,Not)
```

```python
y=4
print(y>3 and y>5)
print(y>3 or y>5)
print(not(y>3 or y>5))
      
```

```python
5.Membership operators(in,not in)
```

```python
text = "Hello my name is Priyanshu."

if "name" in text:
    print("The word name is present in the sentence.")

if "hello" not in text:
    print("The word hello is not present.")
```

```python
6.bitwise logical operators(&,|,!)
```

```python
print(4 & 5)
print(4 | 5)
print(4 ^ 5)
print(~4)
```

```python
7.identity operators(is, is not)
```

```python
a = [1, 2, 3]
b = a
c = [1, 2, 3]

print(a is b)   
print(a is c)    
print(a == c)    

print(a is not c)
```

```python
#simple calculator
num1 = float(input("Enter first number: "))
operator = input("Enter operator (+, -, *, /): ")
num2 = float(input("Enter second number: "))
if operator == '+':
    result = num1 + num2
elif operator == '-':
    result = num1 - num2
elif operator == '*':
    result = num1 * num2
elif operator == '/':
    if num2 != 0:
        result = num1 / num2
    else:
        result = "Error! Division by zero."
else:
    result = "Invalid operator!"
print("Result:", result)
```
